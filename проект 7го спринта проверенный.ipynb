Заменим пропуски в столбцах с рейтингом `Critic_Score` и `User_Score ` на значение-индикатор, к примеру на -1 (отрицательного рейтинга не бывает). Для столбца `Rating` изменим на значение *unknown*.

````python
df_not_null['Critic_Score']=df_not_null['Critic_Score'].fillna(-1)
df_not_null['User_Score']=df_not_null['User_Score'].fillna(-1)
df_not_null['Rating']=df_not_null['Rating'].fillna('unknown')
#Проверяем результат по наличию пропусков
df_not_null.isna().sum()
````

Заменим пропуски в столбцах `EU_sales` и `JP_sales` на среднее значение в зависимости от названия платформы и года выхода игры.
Для этого используем метод *apply()*.

````python
# Напишем функцию для замены пропуска на среднее значение в столбце 'EU_sales':
def fill_sale_EU(row):
    if pd.isna(row['EU_sales']):
        group=df_not_null[(row['Platform']==df_not_null['Platform']) & (row['Year_of_Release']==df_not_null['Year_of_Release'])]
        return group['EU_sales'].mean()
    else:
        return row['EU_sales']
# Заменим пропуски в датафрейме df_not_null:
df_not_null['EU_sales']=df_not_null.apply(fill_sale_EU, axis=1)

# Напишем функцию для замены пропуска на среднее значение в столбце 'JP_sales':
def fill_sale_JP(row):
    if pd.isna(row['JP_sales']):
        group=df_not_null[(row['Platform']==df_not_null['Platform']) & (row['Year_of_Release']==df_not_null['Year_of_Release'])]
        return group['JP_sales'].mean()
    else:
        return row['JP_sales']
# Заменим пропуски в датафрейме df_not_null:
df_not_null['JP_sales']=df_not_null.apply(fill_sale_JP, axis=1)
# Посчитаем количество пропусков в обработанном датафрейме:
import warnings
warnings.filterwarnings('ignore')
df_not_null.isna().sum()
````

### Категоризация данных

Проведем категоризацию данных:
- Разделим все игры по **оценкам пользователей** и выделим такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).
- Разделим все игры по **оценкам критиков** и выделим такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).
Для категоризации данных используем метод *apply(categorize)*, где *categorize* - функция для соотнесения значений ячейки к определенной категории.

````python
# Функция для категоризации
def categorize(value):
    if value<=10.0 and value>=8.0:
        return 'высокая оценка'
    elif value<8.0 and value>3.0:
        return 'средняя оценка'
    elif value<3.0 and value>0.0:
        return 'низкая оценка'
    else:
        return 'без оценки'
# Применим функцию к нашему датафрейму и создаздим новые столбцы 'Оценки критиков' и 'Оценки пользователей'
df_actual['Оценки критиков']=(df_actual['Critic_Score']/10).apply(categorize)
df_actual['Оценки пользователей']=df_actual['User_Score'].apply(categorize)
# Расположим столбцы в удобном порядке с помощью метода reindex()
df_actual=df_actual.reindex(columns=['Name', 'Platform', 'Year_of_Release', 'Genre', 'NA_sales', 'EU_sales', 'JP_sales', 'Other_sales', 'Critic_Score', 'Оценки критиков', 'User_Score', 'Оценки пользователей', 'Rating'])
# Отсортируем значения по оценкам критиков и пользователей в порядке убывания
df_actual.sort_values(by=['Critic_Score', 'User_Score'], ascending=False).head(5)
````

Для наглядности датафрейм был отсортирован по оценкам по убыванию (сначала по оценкам критиков, потом уже пользователей).

````python
# сгруппируем данные по оценкам критиков и скопируем в датафрейм df_grouped_critics
df_grouped_critics=df_actual.groupby(['Оценки критиков'], as_index=False)['Year_of_Release'].count()
# переименуем колонку с подсчетом игр для каждой категории оценки в 'Count_games'
df_grouped_critics=df_grouped_critics.rename(columns={'Year_of_Release':'Count_games'})
# посчитаем процент каждой категории от общего количества
df_grouped_critics['percent_count']=round(df_grouped_critics['Count_games']/df_grouped_critics['Count_games'].sum()*100, 1)
# покажем результат
display(df_grouped_critics.sort_values(['Count_games'], ascending=False))

# сгруппируем данные по оценкам критиков и скопируем в датафрейм df_grouped_users
df_grouped_users=df_actual.groupby(['Оценки пользователей'], as_index=False)['Year_of_Release'].count()
# переименуем колонку с подсчетом игр для каждой категории оценки в 'Count_games'
df_grouped_users=df_grouped_users.rename(columns={'Year_of_Release':'Count_games'})
# посчитаем процент каждой категории от общего количества
df_grouped_users['percent_count']=round(df_grouped_users['Count_games']/df_grouped_users['Count_games'].sum()*100, 1)
# покажем результат
display(df_grouped_users.sort_values(['Count_games'], ascending=False))
````

Как видно, за период с 2000 по 2013 год включительно большая часть игр получила **среднюю** (42.3% и 31.8% от критиков и пользователей соответственно) или **высокую** оценку (17.9% и 13.2%). Доля игр, получившая низкую оценку, от критиков и от пользователей составляет менее 1.0%. Следует учесть, что в выборке находилось около половины игр, по которым не было информации об оценке от критиков и пользователей.
